```python
# Exercício 1

def soma_digitos(n):
    if n == 0:
        return 0
    return n % 10 + soma_digitos(n // 10)

print(soma_digitos(123))
```

```python
# Exercício 2   

def conta_digitos(n):
    if n == 0:
        return 0
    return 1 + conta_digitos(n // 10)

print(conta_digitos(9876))
```

```python
# Exercício 3 

def inverter_string(s):
    if s == "":
        return ""
    return s[-1] + inverter_string(s[:-1])

print(inverter_string("recursão"))
```

```python
# Exercício 4

def eh_palindromo(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return eh_palindromo(s[1:-1])

print(eh_palindromo("arara"))
```

```python
# Exercício 5

def multiplicacao(a, b):
    if b == 0:
        return 0
    if b > 0:
        return a + multiplicacao(a, b - 1)
    return - multiplicacao(a, -b)

print(multiplicacao(3, 4))
```

```python
# Exercícios 6

def maior_lista(lista):
    if len(lista) == 1:
        return lista[0]
    max_restante = maior_lista(lista[1:])
    return lista[0] if lista[0] > max_restante else max_restante

print(maior_lista([3, 7, 2, 9, 5]))
```

```python
# Exercício 7

def mdc(a, b):
    if b == 0:
        return a
    return mdc(b, a % b)

print(mdc(45, 15))
```

```python
#exercícios 8

def contar_elemento(lista, x):
    if not lista:
        return 0
    return (1 if lista[0] == x else 0) + contar_elemento(lista[1:], x)

print(contar_elemento([1, 2, 2, 3, 2], 2))
```

```python
# Exercício 9 

def soma_lista(lista):
    if not lista:
        return 0
    return lista[0] + soma_lista(lista[1:])

print(soma_lista([1, 2, 3, 4,]))
```

```python
# Exercício 10

def eh_primo(n, i=2):
    if n <= 2:
        return n == 2
    if n % i == 0:
        return False
    if i * i > n:
        return True
    return eh_primo(n, i + 1)

print(eh_primo(7))
```
